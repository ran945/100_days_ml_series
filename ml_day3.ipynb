```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("tested.csv")
print(df.shape)
```

```python
df = df.iloc[:, :3]
df
```

 train test split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    df.drop('PassengerId', axis=1),
    df['PassengerId'],
    test_size=0.3,
    random_state=0
)

x_train.shape, x_test.shape
```

standardsacler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit the scaler on the training data
scaler.fit(x_train)

# Transform the training and test data
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
scaler.mean_
```

```python
x_train
```

```python
x_train_scaled
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

why scalling is important

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.scatter(x_train['Survived'],x_train['Pclass'])
ax1.set_title('before scaling')
ax2.scatter(x_train_scaled['Survived'],x_train_scaled['Pclass'],color='red')
ax2.set_title('after scaling')
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
sns.kdeplot(x_train['Survived'],ax=ax1)
sns.kdeplot(x_train['Pclass'],ax=ax1)
ax1.set_title('before scaling')
sns.kdeplot(x_train_scaled['Survived'],ax=ax2)
sns.kdeplot(x_train_scaled['Pclass'],ax=ax2)
ax2.set_title('after scaling')
plt.show()
```
