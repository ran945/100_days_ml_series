```python
import pandas as pd 
import numpy as np 
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv('D:\\dataset\\covid_reviews_50_ordinal.csv')
df
```

```python
df['Age'] = df['Age'].replace('NaN', np.nan)
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    df.drop(columns=['Has_COVID']), 
    df['Has_COVID'], 
    test_size=0.3
)
```

```python
x_test.shape
```

1.Aam zindagi

```python
# adding simple imputer on age column
si =SimpleImputer()
x_train_Age=si.fit_transform(x_train[['Age']])
x_test_Age=si.fit_transform(x_test[['Age']])
```

```python
x_train_Age
```

```python
x_train_Age.shape
```

```python

# ordinal on review column 
oe = OrdinalEncoder(categories=[['Normal', 'Good', 'Best']])
x_train_review = oe.fit_transform(x_train[['Review']])
x_test_review = oe.fit_transform(x_test[['Review']])
```

```python
x_train_review.shape
```

```python
# ohe on gender,city
ohe=OneHotEncoder(drop='first',sparse_output=False)
x_train_gender_city = ohe.fit_transform(x_train[['Gender','City']])
x_test_gender_city = ohe.fit_transform(x_test[['Gender','City']])
x_train_gender_city.shape
```

```python
x_train_gender_city.toarray()
```

```python
x_train_transformed=np.concatenate((x_train_Age,x_train_review,x_train_gender_city),axis=1)
```

```python
x_test_transformed=np.concatenate((x_test_Age,x_test_review,x_test_gender_city),axis=1)
```

```python
from sklearn.compose import ColumnTransformer
transformer=ColumnTransformer(transformers=[
    ('tf1',SimpleImputer(),['Age']),
    ('tf2',OrdinalEncoder(categories=[['Normal','Good','Best']]),['Review']),
    ('tf3',OneHotEncoder(drop='first',sparse_output=False),['Gender','City'])
],remainder='passthrough')
```
